```python
x = df.drop('target', axis=1)
y = df['target']
```

```python
# split the data into train and test set
np.random.seed(42)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
x_train
```

```python
# put models in a dictionary
models = {'logistic_regression':LogisticRegression(),
          'KNN':KNeighborsClassifier(), 
          'Random Forest': RandomForestClassifier()}


# create a function to fit and score the model 
def fit_and_score(models, x_train, x_test, y_train, y_test):
    """
    Fits and evaluates given machine learning models
    models: a dict of different sklearn machine learning models
    x_train: training data(no labels)
    x_test: testing data(no labels)
    y_train: training labels
    y_test: testing labels
    
    """
    # set random seed
    np.random.seed(42)
    # make a dictioary to keep model scores
    model_scores = {}
    # loop through models
    for name, model in models.items():
        # fit the model to the data
        model.fit(x_train, y_train)
        # evaluate the model and appnend its score to the model_scores
        model_scores[name] = model.score(x_test, y_test)
    return model_scores
    
```

```python
model_scores = fit_and_score(models=models, 
                             x_train=x_train,
                             x_test=x_test,
                             y_train=y_train, 
                             y_test=y_test)
model_scores
```

```python
model_compare = pd.DataFrame(model_scores, index=['accuracy'])
model_compare.T.plot.bar()
```

```python
train_scores = []
test_scores = []

# create a list of different values for n_neighbors
neighbors = range(1, 21)

# setup knn instance
knn = KNeighborsClassifier()

# loop through different n_neighbors
for n in neighbors:
    knn.set_params(n_neighbors=n)

    # fit the algorithm
    knn.fit(x_train, y_train)

    # update the training scores list
    train_scores.append(knn.score(x_train, y_train))

    # update the test score list
    test_scores.append(knn.score(x_test, y_test))
max(train_scores), max(test_scores)
train_scores
```

```python
plt.plot(neighbors, train_scores, label='train score')
plt.plot(neighbors, test_scores, label='test score')
plt.xticks(np.arange(1, 21, 1))
plt.xlabel('number of neighbors')
plt.ylabel('model score')
plt.legend()

print(f'maximum knn score on the test data set: {max(test_scores)*100:.2f}')
```

```python
# create a hyperparameter grid for logisticRegression
log_reg_grid = {"C":np.logspace(-4, 4, 20), 
                'solver':['liblinear']}

# create a hyperparameter grid for randomForestClassifier
rf_grid = {"n_estimators":np.arange(10, 1000, 50),
           'max_depth': [None, 3, 5, 10], 
           'min_samples_split':np.arange(2, 20, 2), 
           'min_samples_leaf':np.arange(1, 20, 2)}
```

```python
# tune logistic regression 
np.random.seed(42)

# setup random hyperparameter search for logistic regression 
rs_log_reg = RandomizedSearchCV(LogisticRegression(),
                                param_distributions=log_reg_grid, 
                                cv =5 ,
                                n_iter=20,
                                verbose=True)
# fit random hyperparameter search model for logistic regression
rs_log_reg.fit(x_train, y_train)
```

```python
rs_log_reg.best_params_
```

```python
rs_log_reg.score(x_test, y_test)
```

```python
# set up random seed
np.random.seed(42)

# setup random hyperparameter search for randomForestClassifier
rs_rf = RandomizedSearchCV(RandomForestClassifier(), 
                          param_distributions=rf_grid, 
                          cv=5, 
                          n_iter=20, 
                          verbose=True)
# fit random random hyperparameter search model for random forest
rs_rf.fit(x_train, y_train)
```

```python
rs_rf.best_params_
```

```python
rs_rf.score(x_test, y_test)
```

```python
log_reg_grid = {"C":np.logspace(-4, 4, 30), 
                'solver':['liblinear']}

# setup grid hyperparameter search for logistic regression
gs_lg_rg = GridSearchCV(LogisticRegression(), 
                        param_grid=log_reg_grid, 
                        cv=5,
                        verbose=True)
gs_lg_rg.fit(x_train, y_train)
```

```python
gs_lg_rg.best_params_
```

```python
gs_lg_rg.score(x_test, y_test)
```

```python
# make predictions with tuned model
y_preds = gs_lg_rg.predict(x_test)
y_preds
```

```python
# plot roc curve and calculate AUC metric

y_probs = gs_lg_rg.predict_proba(x_test)[:, 1]

# Compute ROC curve and AUC
fpr, tpr, thresholds = roc_curve(y_test, y_probs)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

```python
# confusion matrix 
print(confusion_matrix(y_test, y_preds))
```

```python
# plot confusion matrix
sns.set(font_scale=1.5)

def plot_conf_mat(y_test, y_preds):
    """
    plots a nice looking confusion matrix using seaborn's heatmap
    """
    fig, ax = plt.subplots(figsize=(3, 3))
    ax = sns.heatmap(confusion_matrix(y_test, y_preds),
                     annot=True, 
                     cbar=False)
    plt.xlabel("true label")
    plt.ylabel("predicted label")

plot_conf_mat(y_test, y_preds)
```

```python
# classification report
print(classification_report(y_test, y_preds))
```
